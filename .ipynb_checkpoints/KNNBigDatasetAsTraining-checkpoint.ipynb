```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path="C:/Users/ABC/Desktop/test1.csv"
```

```python
headernames = ['Age','BusinessTravel','DailyRate','Department','DistanceFromHome','Education' ,'EducationField','EmployeeCount','EmployeeNumber',
             'EnvironmentSatisfaction','HourlyRate','JobInvolvement','JobLevel','JobRole','JobSatisfaction',
             'MonthlyIncome','MonthlyRate','NumCompaniesWorked','PercentSalaryHike','PerformanceRating','RelationshipSatisfaction','StandardHours',
             'StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager','Attrition'
]
```

```python
df=pd.read_csv(path,names=headernames)
```

```python
df
```

```python
 BusinessTravel= {'Travel_Rarely': 1,'Travel_Frequently': 2,'Non-Travel':3}
```

```python
df.BusinessTravel = [BusinessTravel[item] for item in df.BusinessTravel] 
```

```python
df
```

```python
abc=pd.Series(["Sales", "Research & Development", "Human Resources"], dtype="category")
```

```python
abc
```

```python
abc= {"Sales":1, "Research & Development":2, "Human Resources":3}
df.Department= [abc[item] for item in df.Department]
```

```python
df
```

```python
edu= {"Medical":1, "Life Sciences":2, "Other":3,"Marketing":4,"Technical Degree":5,"Human Resources":6}
df.EducationField= [edu[item] for item in df.EducationField]
```

```python
df
```

```python
job= {"Sales Executive":1, "Research Scientist":2, "Laboratory Technician":3,"Manufactoring Director":4,"Healthcare Representative":5,"Manager":6, "Sales Representative":7,"Research Director":8,"Manufacturing Director":9,"Human Resources":10}
df.JobRole= [job[item] for item in df.JobRole]
```

```python
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,30].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.30,random_state=8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=36)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import precision_score
print(precision_score(y_test, y_pred, average="macro"))
```

```python
from sklearn.metrics import recall_score
print(recall_score(y_test, y_pred, average="macro")) 
```
